```python
#creating a User define Function(UDF) to apply to certain column
def clean_text(text):
    for char in text:
        if char in string.punctuation:
            text = text.replace(char,'')
            text = text.strip()
    return text
```

```python
# Now apply clean_text function to text column to remove punctuation
df['clean_text'] = df['text'].apply(lambda x : clean_text(x))

# Apply the clean_text function to the 'section_title' column
df['clean_section_title'] = df['section_title'].apply(lambda x: clean_text(x))

# Apply the clean_text function to the 'article_title' column
df['clean_article_title'] = df['article_title'].apply(lambda x: clean_text(x))


df.head(5)
```

```python
# Importing libraries for stop words removal
import nltk.corpus
from nltk.corpus import stopwords
import gensim
from gensim.utils import simple_preprocess
nltk.download('stopwords')
```

```python
stopwords_list = stopwords.words('english')
stopwords_list
```

```python
#create a custom word list
stopwords_list.extend(['the','that', 'a ','us', 'of', 'in','that', 'made', 'we\'ve', 'after', 'their','his','only', 'shall', 'shallnewlinea','shallnewlinei','newline','whole','unto'])
stopwords_list
```

```python
#Now creating a UDF to remove stopwords from the certain column in dataset.
def stopwords_clean(text):
    word_list = []
    for word in gensim.utils.simple_preprocess(text):
        if word not in stopwords_list:
            word_list.append(word)
    final_list = ' '.join(word_list)
    return final_list
```

```python
df['clean_text'] = df['clean_text'].apply(lambda x : stopwords_clean(x))
df['clean_section_title'] = df['clean_section_title'].apply(lambda x : stopwords_clean(x))
df['clean_article_title'] = df['clean_article_title'].apply(lambda x : stopwords_clean(x))

df.head(5)
```

```python
import gensim
from gensim.models.ldamodel import LdaModel
from gensim import corpora
from gensim.models import CoherenceModel
import numpy as np

# Convert the document-term matrix to a gensim corpus
corpus = gensim.matutils.Sparse2Corpus(dtm.transpose())

# Create a mapping of word IDs to words for LDA
id2word = corpora.Dictionary.from_corpus(corpus, id2word=dict((i, word) for word, i in vectorizer.vocabulary_.items()))

# Define a range of values for k (number of topics)
k_values = range(5,12)

# Lists to store coherence and perplexity values
coherence_values = []
perplexity_values = []

# Iterate over different values of k
for num_topics in k_values:
    lda_model = LdaModel(corpus=corpus, id2word=id2word, num_topics=num_topics, passes=10)

    # Compute topic coherence
    texts = [[id2word[word_id] for word_id, _ in doc] for doc in corpus]
    coherence_model_lda = CoherenceModel(model=lda_model, texts=texts, dictionary=id2word, coherence='c_v')
    coherence_lda = coherence_model_lda.get_coherence()
    coherence_values.append(coherence_lda)

    # Compute perplexity
    perplexity = lda_model.log_perplexity(corpus)
    perplexity_values.append(perplexity)

# Convert lists to numpy arrays for easy analysis
coherence_values = np.array(coherence_values)
perplexity_values = np.array(perplexity_values)

# Print or analyze the coherence and perplexity values as needed
for k, coherence, perplexity in zip(k_values, coherence_values, perplexity_values):
    print(f"K={k}: Coherence={coherence}, Perplexity={perplexity}")
```

```python
import nltk
from gensim.models import Phrases
from gensim.models.phrases import Phraser
from gensim.corpora import Dictionary
from gensim.models import LdaModel

# Sample text data
documents = [
    "machine learning is a subset of artificial intelligence",
    "natural language processing is an essential part of NLP",
    "deep learning models can learn from unstructured data",
    "supervised and unsupervised learning are two types of ML"
]

# Tokenize and preprocess the documents
nltk.download('punkt')
nltk.download('stopwords')
stop_words = set(nltk.corpus.stopwords.words('english'))

tokenized_texts = [nltk.word_tokenize(doc.lower()) for doc in documents]
tokenized_texts = [[token for token in text if token not in stop_words] for text in tokenized_texts]

# Extract bigrams
bigram_model = Phrases(tokenized_texts, min_count=1, threshold=2, delimiter=b'_')
bigram_phraser = Phraser(bigram_model)
bigram_tokenized_texts = [bigram_phraser[doc] for doc in tokenized_texts]

# Create the Dictionary (vocabulary) of words and bigrams
dictionary = Dictionary(bigram_tokenized_texts)

# Convert tokenized texts to bag-of-words (BoW) representation
bow_corpus = [dictionary.doc2bow(doc) for doc in bigram_tokenized_texts]

# Train LDA model
num_topics = 3
lda_model = LdaModel(bow_corpus, num_topics=num_topics, id2word=dictionary, passes=10, random_state=42)

# Print the topics
for idx, topic in lda_model.print_topics():
    print(f"Topic: {idx}")
    print(f"Words: {topic}")
    print()

# Get the topic distribution for a specific document
doc_idx = 0
topic_distribution = lda_model.get_document_topics(bow_corpus[doc_idx])
print(f"Topic distribution for document {doc_idx}: {topic_distribution}")
```

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.feature_extraction.text import TfidfVectorizer

# Assuming you have a list of preprocessed documents called 'documents'
# Create a TF-IDF vectorizer instance
tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')

# Fit and transform the text data to create the TF-IDF matrix
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

# Define the number of topics
num_topics = 30

# Initialize the LSA model
lsa_model_30_topics = TruncatedSVD(n_components=num_topics)

# Fit the LSA model to the TF-IDF matrix
lsa_matrix_30_topics = lsa_model_30_topics.fit_transform(tfidf_matrix)
```

```python
from gensim.models.coherencemodel import CoherenceModel
from gensim.corpora import Dictionary

# Create a Dictionary from your data
dictionary = Dictionary(documents)

# Calculate coherence
coherence_model = CoherenceModel(
    topics=[[word for word, _ in lsa_model_30_topics.show_topic(topic_idx, topn=10)] for topic_idx in range(num_topics)],
    dictionary=dictionary,
    texts=documents,
    coherence='c_v'
)

coherence_score = coherence_model.get_coherence()

print(f"Coherence Score for 30 Topics: {coherence_score}")
```
